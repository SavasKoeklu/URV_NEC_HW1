I executed all 25 Parameters settings in the model_comparision.ipynb, here I recovered all the data for the mape for values here to enhance Analysis. So I have for each of the three datsets a section in this Jupyter notebook

```python
import pandas as pd
import numpy as np

df = pd.read_csv("params.csv")
```

```python
# function for analysis of mape values
def analyze(parameter, mape_values, epoch):
    mape_for_activation = {}
    mape_for_architecture = {}

    for i in range(len(parameter)):
        activation_func = parameter.iloc[i,4]
        architecture = parameter.iloc[i,1]
        if(activation_func not in mape_for_activation):
            mape_for_activation[activation_func] = []
        mape_for_activation[activation_func].append( mape_values[i][epoch] )

        if(architecture not in mape_for_architecture):
            mape_for_architecture[architecture] = []
        mape_for_architecture[architecture].append( mape_values[i][epoch] )


    for activation in mape_for_activation:
        print(f"Average for Activatio function {activation} is:{ np.mean(np.array(mape_for_activation[activation]))}")
    
    
    for archi in mape_for_architecture:
        print(f"Average for Network Architecture {archi} is:{ np.mean(np.array(mape_for_architecture[archi]))}")
```
